```python
#LESSON-2
# Concept 1 — List Comprehensions

# 1. Using a list comprehension, create a list of squares
#    for numbers 1 through 10
squares = [i ** 2 for i in range(1, 11)]
print(squares)

# 2. Using a list comprehension with a condition,
#    filter only the even squares from your list above
even_squares = [s for s in squares if s % 2 == 0]
print(even_squares)

#Concept 2 — Error Handling with try/except

# 3. Use try/except to handle this error gracefully
#    Print a friendly message instead of crashing
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Impossible to divide by 0!!!")

# 4. Use try/except to handle this and print the error message
try:
    number = int("hello")
except Exception as e:
    print(f"Cant convert letters to numbers: {e}")
```

```python
# LESSON-3
# Concept 1 — Reading & Writing Files

# 1. Write a list of 5 city names to a file called "cities.txt"
#    one city per line
with open("cities.txt", "w") as f:
    cities = ["London", "Paris", "Berlin", "Madrid", "Rome"]
    for city in cities:
        f.write(city + "\n")

# 2. Read the file back and print each city in uppercase
with open("cities.txt", "r") as f:
    for line in f:
        print(line.strip().upper())

#Concept 2 — OOP Basics

# 3. Build a class called Dataset with:
#    - attributes: name, rows, columns
#    - method: describe() that prints all three attributes
class Dataset:
    def __init__(self, name, rows, columns):
        self.name = name
        self.rows = rows
        self.columns = columns

    def describe(self):
        print(f"{self.name} has {self.rows} rows and {self.columns} columns.")

dtset = Dataset("Customer_Data", 100, 5)
#print(dtset)
dtset.describe()
```
