```python
#LESSON-2
# 1. Using a list comprehension, create a list of squares
#    for numbers 1 through 10
squares = [i ** 2 for i in range(1, 11)]
print(squares)

# 2. Using a list comprehension with a condition,
#    filter only the even squares from your list above
even_squares = [s for s in squares if s % 2 == 0]
print(even_squares)

# 3. Use try/except to handle this error gracefully
#    Print a friendly message instead of crashing
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Impossible to divide by 0!!!")

# 4. Use try/except to handle this and print the error message
try:
    number = int("hello")
except Exception as e:
    print(f"Cant convert letters to numbers: {e}")
```
