```python
#LESSON-2
# Concept 1 — List Comprehensions

# 1. Using a list comprehension, create a list of squares
#    for numbers 1 through 10
squares = [i ** 2 for i in range(1, 11)]
print(squares)

# 2. Using a list comprehension with a condition,
#    filter only the even squares from your list above
even_squares = [s for s in squares if s % 2 == 0]
print(even_squares)

#Concept 2 — Error Handling with try/except

# 3. Use try/except to handle this error gracefully
#    Print a friendly message instead of crashing
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Impossible to divide by 0!!!")

# 4. Use try/except to handle this and print the error message
try:
    number = int("hello")
except Exception as e:
    print(f"Cant convert letters to numbers: {e}")
```

```python
# LESSON-3
# Concept 1 — Reading & Writing Files

# 1. Write a list of 5 city names to a file called "cities.txt"
#    one city per line
with open("cities.txt", "w") as f:
    cities = ["London", "Paris", "Berlin", "Madrid", "Rome"]
    for city in cities:
        f.write(city + "\n")

# 2. Read the file back and print each city in uppercase
with open("cities.txt", "r") as f:
    for line in f:
        print(line.strip().upper())

#Concept 2 — OOP Basics

# 3. Build a class called Dataset with:
#    - attributes: name, rows, columns
#    - method: describe() that prints all three attributes
class Dataset:
    def __init__(self, name, rows, columns):
        self.name = name
        self.rows = rows
        self.columns = columns

    def describe(self):
        print(f"{self.name} has {self.rows} rows and {self.columns} columns.")

dtset = Dataset("Customer_Data", 100, 5)
#print(dtset)
dtset.describe()
```

```python
## WEEK 1 REVISION

# 1. Create a variable for your age and a variable for your name
my_age = 30
my_name = "Ronit"

# 2. Create a list of 5 countries
countries = ['India', 'U. S. of A.', 'Nepal', 'China', 'Cuba']

# 3. Print the last item in the list using indexing
#    (no hardcoding the index as 4 — think about it)
print(countries[-1])

# 4. Create a dict representing a dataset with keys:
#    "name", "rows", "columns", "source"
data_dict = {"name":"Sales", "rows":100, "columns":12, "source":"Excel"}
print(data_dict)

# 5. Write a function called clean_string that:
#    - takes a string as input
#    - strips whitespace
#    - converts to lowercase
#    - returns the result
def clean_string(strng):
  return strng.strip().lower()
print(clean_string("\n qw d   uhuh  cc\n"))

# 6. Using a list comprehension, create a list of
#    the lengths of each country name from question 2
country_legths = [len(c) for c in countries]
print(country_legths)

# 7. Using a list comprehension with a condition,
#    return only countries whose names are longer than 5 characters
long_countries = [c for c in countries if len(c) > 5]
print(long_countries)

# 8. Write a try/except block that:
#    - tries to access index 10 of your countries list
#    - catches the specific error
#    - prints a friendly message
try:
  print(countries[10])
except Exception as e:
  print(f"Index out of bounds: {e}")

# 9. Write your countries list to a file called "countries.txt"
#    one country per line, then read it back and print each line
with open("countries.txt", "w") as f:
  for c in countries:
    f.write(c + "\n")

with open("countries.txt", "r") as f:
  for line in f:
    print(line.strip())

# 10. Build a class called DataPipeline with:
#     - attributes: name, source, destination
#     - method: describe() that prints all three
#     - method: run() that prints "Running pipeline: {name}"
class DataPipeline:
  def __init__(self, name, source, destination):
    self.name = name
    self.source = source
    self.destination = destination

  def describe(self):
    print(f"Name: {self.name}, Source: {self.source}, Destination: {self.destination}")

  def run(self):
    print(f"Running pipeline: {self.name}")

data_pipeline = DataPipeline("Marketing", "ABC", "XYZ")
data_pipeline.describe()
data_pipeline.run()
```
